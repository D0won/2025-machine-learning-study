### DataFrame.describe()

- 이 메서드는 칼럼별 숫자형 데이터값의 n-percentile 분포도, 평균값, 최댓값, 최솟값을 나타냄
- 이 메서드는 오직 숫자형 칼럼의 분포도만 조사
- 데이터의 분포도를 아는 것은 머신러닝 알고리즘의 성능을 향상시키는 중요한 요소이므로 개략적인 수준의 분포도를 확인할 수 있는 이 메서드는 유용
- 이 메서드는 해당 숫자 칼럼이 숫자형 카테고리 칼럼인지를 판단할 수 있게 도와줌

1. count : 널이 아닌 데이터 건수
2. mean : 전체 데이터의 평균값
3. std : 표준편차
4. min : 최솟값
5. max : 최댓값
6. 25%,50%,75% : n-percentile 값

## DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환

- DataFrame은 기본적으로 파이썬의 리스트, 딕셔너리, 그리고 넘파이 ndarray 등 다양한 데이터로부터 생성될 수 있음
- DataFrame은 반대로 파이썬의 리스트, 딕셔너리, 그리고 넘파이 ndarray 등으로 변환될 수 있음
- 사이킷런은 기본적으로 넘파이 ndarray를 입력 인자로 사용하기에 DataFrame과 넘파이 ndarray 상호 간의 변환은 매우 빈번함

### 넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환하기

- DataFrame은 다른 자료형과 다르게 칼럼명을 갖고 있어 편한 데이터 핸들링이 가능함
- 일반적으로 DataFrame으로 변환 시 칼럼명을 지정해야함
- DataFrame은 기본적으로 행과 열을 가지는 2차원 데이터이기에 2차원 이하의 데이터들만 DataFrame으로 변환 가능함

````python
import numpy as np

col_name1 = ['col1']
list1 = [1,2,3]
array1 = np.array(list1)
print('array1 shape:', array1.shape)

# 리스트를 이용해 DataFrame 생성
df_list1 = pd.DataFrame(list1, columns=col_name1)
print('1차원 리스트로 만든 DataFrame:\n', df_list1)

# 넘파이 ndarray를 이용해 DataFrame 생성
df_array1 = pd.DataFrame(array1, columns=col_name1)
print('1차원 ndarray로 만든 DataFrame:\n', df_array1)
````

````python
col_name2 = ['col1', 'col2', 'col3']

# 2행x3열 형태의 리스트와 ndarray 생성한 뒤 이를 DataFrame으로 변환.
list2 = [[1,2,3],
         [11,12,13]]
array2 = np.array(list2)
print('array2 shape:', array2.shape)

df_list2 = pd.DataFrame(list2, columns=col_name2)
print('2차원 리스트로 만든 DataFrame:\n', df_list2)

df_array2 = pd.DataFrame(array2, columns=col_name2)
print('2차원 ndarray로 만든 DataFrame:\n', df_array2)
````

````python
# key는 문자열 칼럼명으로 매핑, Value는 리스트 형(또는 ndarray) 칼럼 데이터로 매핑
dict = {'col1':[1, 11], 'col2':[2, 22], 'col3':[3, 33]}
df_dict = pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame:\n', df_dict)
````

### DataFrame을 넘파이 ndarray, 리스트, 딕셔너리로 변환하기

- 많은 머신러닝 패키지가 기본 데이터 형으로 넘파이 ndarray를 사용함
- DataFrame을 넘파이 ndarray로 변환하는 것은 DataFrame 객체의 values를 이용해 쉽게 할 수 있음

````python
# DataFrame을 ndarray로 변환
array3 = df_dict.values
print('df_dict.values 타입:', type(array3), 'df_dict.values shape:', array3.shape)
print(array3)
````

````python
# DataFrame을 리스트로 변환
list3 = df_dict.values.tolist()
print('df_dict.values.tolist() 타입:', type(list3))
print(list3)

# DataFrame을 딕셔너리로 변환, to_dict() 매서드 안의 인자를 'list'로 두면 딕셔너리 값이 리스트 형태로 변환
dict3 = df_dict.to_dict('list') 
print('\n df_dict.to_dict() 타입:', type(dict3))
print(dict3)
````

## DataFrame 데이터 삭제

- DataFrame에서 데이터 삭제는 drop() 메서드를 이용함
- drop() 메서드 원형
``` python
DataFrame.drop(labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')
```
- axis : axis=0은 Raw 방향 축, axis=1은 Column 방향 축으로 드롭할 방향을 설정
- labels : 만약 axis=1로 Column 방향 축으로 드롭할 때 레이블 변수에 드롭할 칼럼명을 입력한다. 만약 axis=0으로 Raw 방향 축으로 드롭할 떄 레이블 변수에 드롭할 로우를 입력
- inplace : 만약 inplace = False면 자기 자신의 DataFrame의 데이터는 삭제하지 않으며 삭제된 결과 DataFrame을 반환, 만약 inplace = True면 자신의 DataFrame의 데이터를 삭제한다.

````python
titanic_drop_df = titanic_df.drop('Age_0', axis=1)
titanic_drop_df.head(3)
````

````python
titanic_df.head(3)
````

````python
# default로 inplace = False로 설정되어 있으며 inplace=True로 할 시에 자기 자신의 DataFrame에서 칼럼을 삭제한다.(주의 : inplace=True로 할 경우 반환 값은 None이 된다.)
drop_result = titanic_df.drop(['Age_0', 'Age_by_10', 'Family_No'], axis= 1, inplace=True)
print('inplace=True로 drop 후 반환된 값:', drop_result)
titanic_df.head(3)
````

````python
# axix = 0 으로 설정해 index 0,1,2(맨 위 3개 데이터) 로우를 삭제
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', 15)
print('#### before axis 0 drop ####')
print(titanic_df.head(3))

titanic_df.drop([0,1,2], axis=0, inplace=True)

print('#### after axis 0 drop ####')
print(titanic_df.head(3))
````

## Index 객체

- Index 객체는 RDBMS의 PK와 유사하게 DataFrame, Series의 레코드를 고유하게 식별하는 객체이다.

````python
# 원본 파일 다시 로딩
titanic_df = pd.read_csv('titanic_train.csv')
# index 객체 추출
indexes = titanic_df.index
print(indexes)
# index 객체를 실제 값 array로 변환
print('Index 객체 array 값:\n', indexes.values)
````

````python
# index 객체는 식별성 데이터를 1차원 array로 가지고 있음. 또한 ndarray와 유사하게 단일 값 반환 및 슬라이싱 가능
print(type(indexes.values))
print(indexes.values.shape)
print(indexes[:5].values)
print(indexes.values[:5])
print(indexes[6])
````

````python
# Series 객체는 index 객체를 포함하지만 Series 객체에 연산 함수를 적용할 때 index는 연산에서 제외
series_fair = titanic_df['Fare']
print('Fair Series max 값:', series_fair.max())
print('Fair Series sum 값:', series_fair.sum())
print('sum() Fait Series:', sum(series_fair))
print('Fair Series + 3:\n', (series_fair+3).head(3))
````

````python
# DataFrame 및 Series에 reset_index() 메서드를 수행하면 새롭게 인덱스를 연속 숫자 형으로 할당하며 기존 인덱스는 'index'라는 새로운 칼럼명을 추가
titanic_reset_df = titanic_df.reset_index(inplace=False)
titanic_reset_df.head(3)
````

````python
print('### before reset_index ###')
value_counts = titanic_df['Pclass'].value_counts()
print(value_counts)
print('value_counts 객체 변수 타입:', type(value_counts))
new_value_counts = value_counts.reset_index(inplace=False)
print('### After reset_index ###')
print(new_value_counts)
print('new_value_counts 객체 변수 타입:', type(new_value_counts))
````

- reset_index()의 parameter 중 drop=True로 설정하면 기존 인덱스는 새로운 칼럼으로 추가되지 않고 삭제됨

## 데이터 셀렉션 및 필터링

- 판다스의 경우 iloc[],loc[] 연산자를 통해 단일 값 추출, 슬라이싱, 팬시 인덱싱, 불린 인덱싱을 수행한다.

### DataFrame의 [] 연산자

- DataFrame 바로 뒤에 있는 '[]' 안에 들어갈 수 있는 것은 칼럼명 문자  또는 인덱스로 변환 가능한 표현식이다.

````python
print('단일 칼럼 데이터 추출:\n', titanic_df['Pclass'].head(3))
print('\n여러 칼럼의 데이터 추출:\n', titanic_df[['Survived', 'Pclass']].head(3))
print('[] 안에 숫자 index는 KeyError 오류 발생:\n', "titanic_df[0]")
````

````python
# DataFrame의 [] 내에 판다스의 인덱스 형태로 변환 가능한 표현식은 입력 가능하다.
titanic_df[0:2]
````

````python
# [] 내의 불린 인덱싱 표현도 가능하다.
titanic_df[titanic_df['Pclass'] == 3].head(3)
````

- 정리
    - DataFrame 바로 뒤의 []연산자는 넘파이의 []나 Series의 []와 다름
    - DataFrame 바로 뒤의 [] 내 입력값은 칼럼명을 지정해 칼럼 지정 연산에 사용하거나 불린 인덱스 용도로만 사용해야함.
    - DataFrame[0:2]와 같은 슬라이싱 연산으로 데이터를 추출하는 방법은 사용하지 않는게 좋음.

### DataFrame iloc[] 연산자

- 판다스의 iloc[] 연산자는 위치 기반 인덱싱 방식으로 행과 열 위치를 0을 출발점으로 하는 세로축, 가로축 좌표 정숫값으로 지정하는 방식


````python
data = {'Name' : ['Chulmin', 'Eunkyung', 'Jinwoong', 'Soobeom'],
        'Year' : [2011, 2016, 2015, 2015],
        'Gender' : ['Male', 'Female', 'Male', 'Male']}

# 인덱스를 'one','two','three','four'로 하는 데이터프레임 형성
data_df = pd.DataFrame(data, index=['one', 'two', 'three', 'four'])
data_df
````

````python
# iloc[행 위치 정숫값, 열 위치 정숫값]처럼 DataFrame의 행 위치와 열 위치를 좌표 정숫값 형태로 입력하여 해당 위치에 있는 데이터를 가져올 수 있음
data_df.iloc[0, 0]
````

````python
# 반면에 iloc[]에 DataFrame의 인덱스 값이나 칼럼명을 입력하면 오류 발생
data_df.iloc[0, 'Name']
````

````python
# iloc[]은 열 위치에 -1을 입력하여 DataFrame의 가장 마지막 열 데이터를 가져오는 데 자주 사용 특히 맨 마지막 칼럼이 타깃 값인 경우 많이 사용
print("\n 맨 마지막 칼럼 데이터 [:, -1] \n", data_df.iloc[:, -1])
print("\n 맨 마지막 칼럼을 제외한 모든 데이터 [:, :-1] \n", data_df.iloc[:, :-1])
````

- iloc[]는 슬라이싱과 팬시 인덱시은 제공하나 명확한 위치 기반 인덱싱이 사용되어야 하는 제약으로 인해 불린 인덱싱은 제공하지 않는다.

### DataFrame loc[] 연산자

- loc[]는 명칭 기반 인덱싱으로 데이터 프레임의 인덱스 값으로 행 위치를, 칼럼의 명칭으로 열 위치를 지정하는 방식

````python
data_df.loc['one', 'Name']
````

````python
# 첫번째 행 위치에 무턱대고 정숫값을 입력해서는 안됨
# data_df.loc[0, 'Name']
````

````python
# loc[]에 슬라이싱 기호를 적용하면 종료값 - 1까지가 아닌 종료 값까지 포함함
print('명칭 기반 loc slicing\n', data_df.loc['one':'two', 'Name'], '\n')
# 비교
print('위치 기반 iloc slicing\n', data_df.iloc[0:1, 0])
````

### 불린 인덱싱

- 판다스의 DataFrame에서는 []와 loc[]에서 불린 인덱싱을 지원한다.

````python
# 새로 불러옴
titanic_df = pd.read_csv('titanic_train.csv')
# 나이가 60 이상인 사람들만 추출
titanic_boolean = titanic_df[titanic_df['Age'] > 60]
print(type(titanic_boolean))
titanic_boolean
````

````python
# 나이가 60 이상인 사람들의 이름과 나이만 추출
titanic_df[titanic_df['Age'] > 60][['Name','Age']].head(3)
````

````python
# loc[] 이용
titanic_df.loc[titanic_df['Age'] > 60, ['Name', 'Age']].head(3)
````

````python
# AND 연산자(&),OR 연산자(|), Not 연산자(~) 사용 가능
titanic_df[(titanic_df['Age'] > 60) & (titanic_df['Pclass']==1) &
           (titanic_df['Sex']=='female')]
````

````python
# 개별 조건을변수에 할당 후 변수 결합하여 불린 인덱싱 수행
cond1 = titanic_df['Age'] > 60
cond2 = titanic_df['Pclass'] == 1
cond3 = titanic_df['Sex'] == 'female'
titanic_df[cond1 & cond2 & cond3]
````
