## 핵심 객체
- 판다스의 핵심 객체는 DataFrame이다.
    - DataFrame이란 여러 개의 행과 열로 이뤄진 2차원 데이터를 담는 데이터 구조체이다.
- 다른 중요 객체로 Index와 Series이다.
    - Index는 RDBMS의 PK처럼 개별 데이터를 고유하게 식별하는 Key값이다.
    - Series는 DataFrame과 다리 칼럼이 하나뿐인 데이터 구조체이다. -> DataFrame은 여러 개의 Series로 이뤄졌다고 할 수 있다.


## 판다스 시작 - 파일을 DataFrame으로 로딩, 기본 API

````python
import pandas as pd
````

### read_csv()

- read_csv()는 호출 시 파일명 인자로 들어온 파일을 로딩해 DataFrame 객체로 반환

- read_csv()로 받아온 DataFrame의 맨 왼쪽 칼럼은 판다스의 Index 객체값임

````python
# read_csv('파일경로')
titanic_df = pd.read_csv('titanic_train.csv')
print('titanic 변수 type:', type(titanic_df))
titanic_df
````

````python
# DataFrame.head()는 DataFrame의 맨 앞에 있는 N개의 Raw를 반환
titanic_df.head(3) 
````

````python
# DataFrame.shape는 DataFrame의 행과 열 크기를 알아보는 변수
titanic_df.shape
````

````python
# DataFrame.info()는 총 데이터 건수와 데이터 타입, Null 데이터 건수를 알 수 있음
titanic_df.info()
````

### DataFrame.describe()

- 이 메서드는 칼럼별 숫자형 데이터값의 n-percentile 분포도, 평균값, 최댓값, 최솟값을 나타냄
- 이 메서드는 오직 숫자형 칼럼의 분포도만 조사
- 데이터의 분포도를 아는 것은 머신러닝 알고리즘의 성능을 향상시키는 중요한 요소이므로 개략적인 수준의 분포도를 확인할 수 있는 이 메서드는 유용
- 이 메서드는 해당 숫자 칼럼이 숫자형 카테고리 칼럼인지를 판단할 수 있게 도와줌

1. count : 널이 아닌 데이터 건수
2. mean : 전체 데이터의 평균값
3. std : 표준편차
4. min : 최솟값
5. max : 최댓값
6. 25%,50%,75% : n-percentile 값

````python
titanic_df.describe()
````

````python
# value_counts() 메서드는 해당 칼럼값의 유형과 건수를 확인할 수 있음
value_counts = titanic_df['Pclass'].value_counts()
print(type(value_counts))
print(value_counts)
````

````python
# DataFrame의 [] 연산자 내부에 칼럼명을 입력하면 해당 칼럼에 해당하는 Series 객체를 반환
titanic_pclass = titanic_df['Pclass']
print(type(titanic_pclass))
````

````python
# Series 객체는 Index와 단 하나의 칼럼으로 구성된 데이터 세트이다.
titanic_pclass.head()
````

````python
print('titanic_df 데이터 건수:', titanic_df.shape[0])

# value_counts()는 Null값을 포함하여 개별 데이터 값의 건수를 계산할지를 dropna 인자로 판단
# value_counts()는 디폴트로 dropna=True이므로 value_counts(dropna=True)와 동일
print(titanic_df['Embarked'].value_counts())
print(titanic_df['Embarked'].value_counts(dropna=False)) 
````

## DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환

- DataFrame은 기본적으로 파이썬의 리스트, 딕셔너리, 그리고 넘파이 ndarray 등 다양한 데이터로부터 생성될 수 있음
- DataFrame은 반대로 파이썬의 리스트, 딕셔너리, 그리고 넘파이 ndarray 등으로 변환될 수 있음
- 사이킷런은 기본적으로 넘파이 ndarray를 입력 인자로 사용하기에 DataFrame과 넘파이 ndarray 상호 간의 변환은 매우 빈번함

### 넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환하기

- DataFrame은 다른 자료형과 다르게 칼럼명을 갖고 있어 편한 데이터 핸들링이 가능함
- 일반적으로 DataFrame으로 변환 시 칼럼명을 지정해야함
- DataFrame은 기본적으로 행과 열을 가지는 2차원 데이터이기에 2차원 이하의 데이터들만 DataFrame으로 변환 가능함

````python
import numpy as np

col_name1 = ['col1']
list1 = [1,2,3]
array1 = np.array(list1)
print('array1 shape:', array1.shape)

# 리스트를 이용해 DataFrame 생성
df_list1 = pd.DataFrame(list1, columns=col_name1)
print('1차원 리스트로 만든 DataFrame:\n', df_list1)

# 넘파이 ndarray를 이용해 DataFrame 생성
df_array1 = pd.DataFrame(array1, columns=col_name1)
print('1차원 ndarray로 만든 DataFrame:\n', df_array1)
````

````python
col_name2 = ['col1', 'col2', 'col3']

# 2행x3열 형태의 리스트와 ndarray 생성한 뒤 이를 DataFrame으로 변환.
list2 = [[1,2,3],
         [11,12,13]]
array2 = np.array(list2)
print('array2 shape:', array2.shape)

df_list2 = pd.DataFrame(list2, columns=col_name2)
print('2차원 리스트로 만든 DataFrame:\n', df_list2)

df_array2 = pd.DataFrame(array2, columns=col_name2)
print('2차원 ndarray로 만든 DataFrame:\n', df_array2)
````

````python
# key는 문자열 칼럼명으로 매핑, Value는 리스트 형(또는 ndarray) 칼럼 데이터로 매핑
dict = {'col1':[1, 11], 'col2':[2, 22], 'col3':[3, 33]}
df_dict = pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame:\n', df_dict)
````

### DataFrame을 넘파이 ndarray, 리스트, 딕셔너리로 변환하기

- 많은 머신러닝 패키지가 기본 데이터 형으로 넘파이 ndarray를 사용함
- DataFrame을 넘파이 ndarray로 변환하는 것은 DataFrame 객체의 values를 이용해 쉽게 할 수 있음

````python
# DataFrame을 ndarray로 변환
array3 = df_dict.values
print('df_dict.values 타입:', type(array3), 'df_dict.values shape:', array3.shape)
print(array3)
````

````python
# DataFrame을 리스트로 변환
list3 = df_dict.values.tolist()
print('df_dict.values.tolist() 타입:', type(list3))
print(list3)

# DataFrame을 딕셔너리로 변환, to_dict() 매서드 안의 인자를 'list'로 두면 딕셔너리 값이 리스트 형태로 변환
dict3 = df_dict.to_dict('list') 
print('\n df_dict.to_dict() 타입:', type(dict3))
print(dict3)
````

## DataFrame의 칼럼 데이터 세트 생성과 수정

- DataFrame의 칼럼 데이터 세트 생성과 수정 역시 [] 연산자를 이용해 쉽게 가능

````python
# 타이타닉 데이터프레임에 새로운 칼럼 추가
titanic_df['Age_0'] = 0
titanic_df.head(3)
````

````python
# 기존 칼럼 Series를 가공해 새로운 칼럼 Series인 Age_by_10과 Family_No가 새롭게 DataFrame에 추가됨
titanic_df['Age_by_10'] = titanic_df['Age'] * 10
titanic_df['Family_No'] = titanic_df['SibSp'] + titanic_df['Parch'] + 1
titanic_df.head(3)
````

````python
# 새로운 칼럼 'Age_by_10'에 일괄적으로 기존값 + 100으로 업데이트
titanic_df['Age_by_10'] = titanic_df['Age_by_10'] + 100
titanic_df.head(3)
````

## DataFrame 데이터 삭제

- DataFrame에서 데이터 삭제는 drop() 메서드를 이용함
- drop() 메서드 원형
``` python
DataFrame.drop(labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')
```
- axis : axis=0은 Raw 방향 축, axis=1은 Column 방향 축으로 드롭할 방향을 설정한다.
- labels : 만약 axis=1로 Column 방향 축으로 드롭할 때 레이블 변수에 드롭할 칼럼명을 입력한다. 만약 axis=0으로 Raw 방향 축으로 드롭할 떄 레이블 변수에 드롭할 로우를 입력한다.

````python
titanic_drop_df = titanic_df.drop('Age_0', axis=1)
titanic_drop_df.head(3)
````
